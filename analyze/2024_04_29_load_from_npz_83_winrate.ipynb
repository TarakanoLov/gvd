```python
import numpy as np
import pickle
```

```python
x_train = np.load('../ml/train_data/current_data/compressed_x_train2400.npy.npz')
x_train = x_train.f.arr_0
y_train = np.load('../ml/train_data/current_data/compressed_y_train2400.npy.npz')
y_train = y_train.f.arr_0

x_test = np.load('../ml/train_data/current_data/compressed_x_train1200.npy.npz')
x_test = x_test.f.arr_0
y_test = np.load('../ml/train_data/current_data/compressed_y_train1200.npy.npz')
y_test = y_test.f.arr_0
```

```python
with open(f'../ml/train_data/current_data/x_train.pkl', 'rb') as f:
    x_train = np.concatenate([x_train, pickle.load(f)], axis=0)
with open(f'../ml/train_data/current_data/y_train.pkl', 'rb') as f:
    y_train = np.concatenate([y_train, pickle.load(f)], axis=0)
```

```python
y_test
```

```python
import lightgbm as lgb

train_data = lgb.Dataset(x_train, label=y_train)
 
# Create a LightGBM dataset for testing with features X_val and labels Y_val,
# and specify the reference dataset as train_data for consistent evaluation
test_data = lgb.Dataset(x_test, label=y_test, reference=train_data)
```

```python
params = {
    'objective': 'regression',
    'metric': 'mse',
    'learning_rate': 0.1
}
# 0.137879
num_round = 10000
bst = lgb.train(params, train_data, num_round, valid_sets=[
                test_data], early_stopping_rounds=50)
```
